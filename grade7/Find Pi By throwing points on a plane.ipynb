## First: finding probabilities by throwing random points on a plane
Imagine you have a perfectly square piece divided into 4 equal sections. If you throw grains of sand down on it you'd expect about a quarter of the sand to land in the first section, a quarter to land in the second section, a quarter to land in the third, and a quarter to land in the fourth section.  Lets try this out.

```python
import random
from matplotlib import pyplot as plt
import numpy as np

def generateGrainOfSand():
    '''Returns a grain of sand with a x coordinate between 0 and 1 and a y coordinate between 0 and 1 '''
    
    # random.random() generates a number between 0 and 1 at random
    x = random.random()
    y = random.random()
    return (x,y)
```

```python
generateGrainOfSand()
```

So our function works to find a random grain of sand. Now, lets generate 10 of them and plot them on a piece of paper.

```python
x = [] # empty array that we'll use to hold x and y values
y = []

# call grain of sand 100 times
for i in range(100):
    dat = generateGrainOfSand()
    x.append(dat[0])
    y.append(dat[1])

# plt.scatter will plot x,y points on a graph
plt.scatter(x=x, y=y, marker='o')
```

## Excercise 1: Write a function that throws a thousand grains of sand into a square and plot the results

```python
import random
from matplotlib import pyplot as plt
import numpy as np

def generateGrainOfSand():
    '''Returns a grain of sand with a x coordinate between 0 and 1 and a y coordinate between 0 and 1 '''
    # random.random() generates a number between 0 and 1 at random
    x = random.random()
    y = random.random()
    return (x,y)

x = [] # empty array that we'll use to hold x and y values
y = []

# call grain of sand 100 times
for i in range(1000):
    dat = generateGrainOfSand()
    x.append(dat[0])
    y.append(dat[1])

# plt.scatter will plot x,y points on a graph
plt.scatter(x=x, y=y, marker='o')
```

If you count the above you'll find that about the number of points in each quadrant is about 25% or the total, or $\frac{1}{4}$. 

Throwing random points at a plane and counting the fraction of points that fall in a specific spot is a useful way to calculate difficult to find values. For example, the number $\pi$.  

If you recall, the area of a circle is given by $\pi r^2$, where is the radius of the circle. this equation is easy if the radius is 1, and $A=\pi$.  We'll plot a quarter circle as follows.

You'll need to know the equation of a circle of radius 1 is 
$$ x^2 + y^2 = 1 $$

From that you can solve for $y$ and get that 
$$ y = \sqrt{1-x^2}$$


```python
# the function np.linspace(0,1) generates numbers evenly spaced between 0 and 1.  
x = np.linspace(0,1)

# We're going to find y, given the equation above, for every value of x #
y = np.sqrt(1-x**2)
```

```python
plt.plot(x,y)
```

Now for the cool part: if you randomly throw point into that square, the probability that a point lands under the curve is equal to the fraction of the area under the curve. But we know the area, because its one quarter of a circle with radius 1.  In other words, the area under the circle is
$$ A_{\text{Under Circle}} = \frac{1}{4}\pi $$

So if we throw lots of sand, 
$$ 4*\frac{\text{number of grains of sand that falls under the circle}}{\text{total grains of sand}} \approx \pi$$


So here's how you're going to find pi: 
1. Generate a random x,y pair
2. if $y>\sqrt{(1-x)}$, its outside of the circle. Otherwise its inside the circle. 
3. Count all the times its inside the circle. 
4. after a large number of counts, find $\pi$ by calculating  $4*\frac{\text{number of grains of sand that falls under the circle}}{\text{total grains of sand}} \approx \pi$
